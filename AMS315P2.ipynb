```r
data <- read.csv("P2_378631.csv",header=TRUE)
data
summary(data)
```

```r
ME <- lm(Y ~ E1+E2+E3+E4, data = data)
summary(ME)
```

```r
ME_full <- lm(Y ~ (E1+E2+E3+E4+G1+G2+G3+G4+G5+G6+G7+G8+G9+G10+G11+G12+G13+G14+G15+G16+G17+G18+G19+G20)^2, data = data)
summary(ME_full)
```

```r
plot(resid(ME_full) ~ fitted(ME_full), main = 'Residual plot for all combo')
```

```r
library(MASS)
transformation <- boxcox(ME_full)
optimalLambda <- 1.5
optimalLambda
```

```r
ME_trans1 <- lm(Y^optimalLambda ~ (E1+E2+E3+E4+G1+G2+G3+G4+G5+G6+G7+G8+G9+G10+G11+G12+G13+G14+G15+G16+G17+G18+G19+G20)^2, data = data)
summary(ME_trans1)
```

```r
plot(resid(ME_trans1) ~ fitted(ME_trans1), main = 'Residual plot for all combo after boxcox')
```

```r
# install.packages('IRkernel')
# IRkernel::installspec(name = 'ir35', displayname = 'R 4.0.0')
install.packages('leaps', dependencies=TRUE, repos='http://cran.rstudio.com/')
library(leaps)
MF <- regsubsets(model.matrix(ME_trans1)[,-1], I((data$Y)^1.5),
                nbest = 1, nvmax = 5,
                method = 'forward', intercept = TRUE)
temp <- summary(MF)
```

```r
install.packages('knitr', dependencies=TRUE, repos='http://cran.rstudio.com/')
library(knitr)
```

```r
var <- colnames(model.matrix(ME_trans1))
M_test <- apply(temp$which, 1, 
               function(x) paste0(var[x], collapse='+'))
kable(data.frame(cbind(model = M_test, adjR2 = temp$adjr2, BIC = temp$bic)),
     caption = 'test model summary')
```
